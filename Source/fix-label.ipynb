```python
from huggingface_hub import HfApi
import os
# Đẩy lên huggingface
def push(secret_value_0=None, path="GEIO.json"):
    if not secret_value_0:
        secret_value_0 = os.getenv("HF_TOKEN")
    if not secret_value_0:
        raise ValueError("Missing HF token. Set HF_TOKEN env var or pass token.")
    output_dir = "/kaggle/working/fixing.jsonl"
    api = HfApi(token=secret_value_0)
    api.upload_file(
        path_or_fileobj=output_dir,  # Đường dẫn tới file
        path_in_repo=path,  # Tên file trong Hugging Face
        repo_id="PhongGoldFish/refactor3000",  # Repo ID của bạn trên Hugging Face
        repo_type="dataset",  # Loại repo (dataset)
    )
    print("Đã xử lý và ghi lại file.")
push()
```

```python
def check_repeat(text):
    dict_layer = {}
    for i in text.split():
        #print(i)
        if i not in dict_layer:
            dict_layer[i] = 0
        if i in dict_layer:
            dict_layer[i] += 1
            if dict_layer[i] >= 15:
                return False
    return True

check_repeat("dscihcuis")
```

```python
from datasets import load_dataset
import json

output_dir = "/kaggle/working/output.jsonl"
my_dataset = load_dataset("PhongGoldFish/refactor3000",split="train")
count = 0
with open(output_dir,"w") as f:
    for item in my_dataset:
        if not(check_repeat(item["instruction"]) and check_repeat(item["input"]) and check_repeat(item["output"])):
             f.write(json.dumps([item,count], ensure_ascii=False) + '\n')
        count+=1
print("Done!")
```

```python
from datasets import load_dataset
import json
input_dir = "/kaggle/input/fushion/fushion.jsonl"
# len input_dir = 1955
my_dataset = load_dataset("PhongGoldFish/chatDoctor",split="train")
output_dir = "/kaggle/working/fixing.jsonl"
number = []
input_load = []
length = len(my_dataset)
with open(input_dir,"r",encoding="utf-8") as f1,open(output_dir,"w",encoding="utf-8") as f2:
    for item in f1:
        data = json.loads(item)
        nums = int(data[1])
        number.append(nums)
        input_load.append(data)
    for i in range(length):
        if i not in number:
            f2.write(json.dumps(my_dataset[i],ensure_ascii=False)+"\n")
        else:
            for item in input_load:
                if item[1] == i:
                    f2.write(json.dumps(item[0],ensure_ascii=False)+"\n")
            
                
    
```
